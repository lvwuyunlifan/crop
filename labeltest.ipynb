```python
val_list = np.array(label_test["label"]).tolist()
test_list = np.array(result["probability"]).tolist()
print("val_list 10: ", val_list[0:10])
print("test_list 10: ", test_list[0:10])
    
```

```python
lens = len(val_list)

right_list = []
label_list = []
tomato = 0
mean_tomato = 0
mean_all = 0
for j in range(61):
    numbers = 0
    right = 0
    
    for i in range(lens):
        if int(val_list[i]) == j:
            if test_list[i] != val_list[i]:
                print('    columns {0} label is {1}, but test is {2}'
                      .format(i, val_list[i], test_list[i]))
            else:
                right += 1
                if 40<j:
                    tomato += 1
                
            numbers += 1
    
    mean_all += round((100*right/numbers), 2)
    mean_tomato += round((100*tomato/numbers), 2)
    right_list.append(round((100*right/numbers), 2))
    label_list.append(j)
    print("----------------")
    print('label {0} number = {1}, accuracy = {2}%'.format(j, numbers, round((100*right/numbers), 2)))
    print("----------------")
    
print('mean_all is {}'.format(mean_all/61))
print('Tomato mean_all is {}'.format(tomato/20))
```
